```python
import PyPDF2
import spacy
from spacy.matcher import Matcher
import pdftotext
import pandas as pd
import os
```

```python
! python -m spacy download en_core_web_sm
```

```python
path='/content/DATASET' #file path for corpus
files = [os.path.join (path, f) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
pdfs = []
for i in files:
    pdfs.append(i)

merger = PyPDF2.PdfFileMerger(strict=False)

for pdf in pdfs:
    merger.append(pdf)

merger.write("final.pdf")
```

```python
with open("final.pdf", "rb") as f:
    pdf = pdftotext.PDF(f)
```

```python
finalpg=[]
for page in pdf:
    pg=page.strip()
    finalpg.append(pg)
print(finalpg[0])
```

```python
# load pre-trained model
nlp = spacy.load('en_core_web_sm')

# initialize matcher with a vocab
matcher = Matcher(nlp.vocab, validate=True)

def extract_name(resume_text):
    nlp_text = nlp(resume_text)
    
    # First name and Last name are always Proper Nouns
    pattern = [{'POS': 'PROPN'}, {'POS': 'PROPN'}]
    
    matcher.add('NAME', None, [pattern])
    
    matches = matcher(nlp_text)
    
    for match_id, start, end in matches:
        span = nlp_text[start:end]
        return span.text
```

```python
extract_name(finalpg[0])
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(finalpg[0])


for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
nlp = spacy.load('en_core_web_sm')

def extract_hobbies(resume_text):
  nlp_text = nlp(resume_text)

  tokens = [token.text for token in nlp_text if not token.is_stop]
  data = pd.read_csv("/content/DATASET/PriyankaRaikwarRESUME.pdf")

  skills = list(data.columns.values)
    
  skillset = []

  for token in tokens:
        if token.lower() in skills:
            skillset.append(token)

  return [i.capitalize() for i in set([i.lower() for i in skillset])]
```

```python
if __name__ == '__main__':
    directory = '/content/DATASET'

    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        text = finalpg[0]
        skill = extract_hobbies(text)

        print(f," : ",skill)
```
